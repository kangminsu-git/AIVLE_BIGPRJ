```python
import pandas as pd
```

```python
user = pd.read_csv('./data/users.csv', encoding='utf-8')
user.head(10)
```

```python
solved_problem = pd.DataFrame()
solved_problem['handle'] = user['handle']
solved_problem
```

```python
solved_temp = pd.DataFrame()
for i in range(1,41):
    temp = pd.read_csv(f'./data/solved_problem/solved{i}.csv')
    solved_temp = pd.concat([solved_temp, temp])
solved_temp.reset_index(inplace=True, drop=True)
solved_temp
```

```python
solved_problem['solved_problem'] = solved_temp['solved_problem']
solved_problem
```

```python
solved_problem.to_csv('./data/solved_problems.csv', index=False)
```
