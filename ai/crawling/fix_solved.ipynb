```python
import pandas as pd
```

```python
solved_problem = pd.read_csv('./data/solved_problems.csv')
solved_problem
```

```python
error_user = pd.read_pickle('./data/error.pkl')
error_user
```

```python
error_solved = pd.DataFrame()
error_solved.index = error_user.index
error_solved['handle'] = error_user
error_solved
```

```python
error_problem = pd.read_csv('./data/solved_problem/solved_error.csv')
error_problem2 = pd.read_csv('./data/solved_problem/solved_error2.csv')
error_problem = pd.concat([error_problem, error_problem2]).reset_index(drop=True)
error_problem['solved_problem'].explode()
```

```python
error_solved['solved_problem'] = list(error_problem['solved_problem'].explode())
error_solved
```

```python
for i in error_solved.index:
    solved_problem.loc[i,'solved_problem'] = error_solved['solved_problem'][i]

solved_problem.loc[solved_problem.index.isin(error_solved.index)]
```

```python
solved_problem.to_csv('./data/solved_problems.csv', index=False)
```
